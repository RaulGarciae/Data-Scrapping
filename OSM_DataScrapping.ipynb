```python
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
import osmnx as ox
import matplotlib.pyplot as plt
```

```python
# Define the extent using a shapefile
extent_shp_path = 'C:/Users/garciaes/Desktop/Working/PythonTry/poland.shp'
gdf = gpd.read_file(extent_shp_path)
```

```python
features = ox.features_from_polygon(gdf.geometry.values[0], tags={'amenity': 'fuel'})
```

```python
# Convert the features to a GeoDataFrame
fuel = gpd.GeoDataFrame(features)
```

```python
# Keep only desired geometry type
fuel_p= fuel[fuel.geometry.type == 'Point']
```
